#### Let us import all the libraries and it's methods/functions that we need

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import time
import warnings

warnings.filterwarnings('ignore')

from sklearn import preprocessing
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import roc_curve, roc_auc_score, accuracy_score, confusion_matrix
```

```python
# Read the dataset

df= pd.read_csv('lymph_dataset.csv')
```

```python
# Perform label encoding on non numerical features (categorical features)
# Why? Because machine learning models work on numbers and use numbers to perform calculations. Therefore, we
# need to map the categorical features to numerical values

le = preprocessing.LabelEncoder()

for column_name in df.columns:
    if df[column_name].dtype == object:
        df[column_name] = le.fit_transform(df[column_name])
    else:
        pass
```

```python
"""
Let us now visualize the column 'class'
We can clearly see that the classes with values 1 and 2 have large instances while the classes with values
0 and 3 barely have few instances. This is a perfect example of imbalanced class dataset.
Techniques to over come this problem are:

1. Over sampling of minority classes (0 and 3): In this method we generate similar instances of entries having
   class values 0 and 3 to match the number of instances of classes 1 and 2 (these are majority classes).
   
2. SMOTE Analysis: In SMOTE analysis, an algorithm learns the probability distribution of the minority classes
   and then generates new samples that match the number of samples of classes 1 and 2.

In fact, we're also not going to perform feature engineering.
We're going to keep this simple and we're not going to try any of these for now.

"""

sns.set(style="darkgrid")
ax = sns.countplot(df['class'], label="Count")
```

```python
X = df.drop('class', axis=1)  
y = df['class']
```

```python
"""
Plotting the correlation matrix

A correlation matrix tells us how strong two features are related. A correlation value of 1 indicates that 
two features are proportional. A correlation value of -1 indicates that the two features are inversely 
proportional.

From the below graph we can see that the columns regeneration_of and lymph_nodes_dimin are almost related.
We can remove one of the two since there's a good correlation between them.

"""
plt.figure(figsize= (20, 20))
sns.heatmap(X.corr(), cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10})
```

```python
# We drop the regeneration_of column and keep the lymph_nodes_dimin column

X= X.drop('regeneration_of', axis= 1)
```

```python
"""
Perform splitting of the dataset. Here we use 75% of the entries for training the data and 25% for
testing the data. We set a random state so that the dataset is split in the same way when this code is 
transferred to another computer and we also gain same results.
"""

X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.25, random_state= 42)

"""
Here we perform feature scaling. Feature scaling means adjusting data that has different scales so as to 
avoid biases when the machine learning model learns. What this means in simple terms is:

Assume feature 1 has values between 0 and 100 while feature 2 has values between 0 and 50000.
Feature scaling brings the values of the two features within a same range, for ex: [-1, 1]
"""
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
"""
Hyper Parameter tuning:

An important stage in Machine Learning is to choose which model and what all parameters for the model?
So what we do is to choose a few models and then few parameters for the model and perform a technique called
Grid Search.

The Grid Search technique, determines the best parameters for each model and also returns it's scores. The way,
it determines the best parameter is by combination of each parameter with another.

For ex: Take an example of kNN model and we have 2 parameters: Metric and n_neighbors. Grid Search runs the model
for the values:
(3, minkowski), (4, minkowski), (5, minkowski)
(3, manhattan), (4, manhattan), (5, manhattan)
(3, euclidean), (4, euclidean), (5, euclidean)

So we have nine possibilities and Grid Search trains the kNN model 9 times and then compares their scores and
returns the best.

This happens for every other model that we have defined below.

"""

from sklearn.model_selection import GridSearchCV

model_params= {
    'knn': {
        'model': KNeighborsClassifier(),
        'params': {
            'n_neighbors': [3, 4, 5],
            'metric': ['minkowski', 'manhattan', 'euclidean']
        }
    },
    'rfc': {
        'model': RandomForestClassifier(),
        'params': {
            'n_estimators': [100, 150, 200],
            'criterion': ['gini', 'entropy'],
            'max_depth': [5, 6, 7]
        }
    },
    'gnb': {
        'model': GaussianNB(),
        'params': {
            'var_smoothing': [1e-8, 1e-9, 1e-10]
        }
    },
    'svc': {
        'model': SVC(),
        'params': {
            'kernel': ['rbf', 'sigmoid'],
            'C': [0.5, 1, 1.5],
            'gamma': ['scale', 'auto']
        }
    },
    'dtc': {
        'model': DecisionTreeClassifier(),
        'params': {
            'criterion': ['gini', 'entropy'],
            'splitter': ['random', 'best']
        }
    }
}
```

```python
"""
Here we run the model for each combination of the parameters and then find the best parameter combination for
each model
"""

scores= []

for model_name, mp in model_params.items():
    t1= time.time()
    clf= GridSearchCV(mp['model'], mp['params'], cv= 5, return_train_score= False)
    clf.fit(X_train_scaled, y_train)
    scores.append({
        'model': model_name,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_
    })
    t2= time.time()
    print('Time for ', model_name, ' model : ', (t2-t1), ' seconds')
    
pd.set_option('display.max_colwidth', -1)
df= pd.DataFrame(scores)
df
```

```python
"""
Above, we can see that the Random Forest Classifier has performed the best with the hyper-parameters:
gini as criterion, max_depth as 6 and n_estimators as 150.
"""

"""
Now let us compare few models based on the metric F1 score. Models will be initialized with their 
default hyperparameters.
"""

knn= KNeighborsClassifier()
rf= RandomForestClassifier()
dtc= DecisionTreeClassifier()
gnb= GaussianNB()
lr= LogisticRegression()
```

```python
for clf, label in zip([knn, rf, dtc, gnb], ['KNN', 'RF', 'DTC', 'GNB']):
    scores= cross_val_score(clf, X_train_scaled, y_train, cv= 5, scoring= 'f1_macro')
    print('F1 scores: %0.9f (+/- %0.2f) [%s] Model'%(scores.mean(), scores.std(), label))
```

```python
"""
From the above, we see that the Random Forest Classifier has the best F1 score. So, we choose that 
for training.
"""

rfc= RandomForestClassifier()
rfc.fit(X_train_scaled, y_train)
rfc_y_pred= rfc.predict(X_test_scaled)
```

```python
"""
Let us plot the confusion matrix for the Random Forest Classifier
"""

data = {'y_Actual': y_test, 'y_Predicted': rfc_y_pred }

df = pd.DataFrame(data, columns=['y_Actual','y_Predicted'])
cm_rfc = pd.crosstab(df['y_Actual'], df['y_Predicted'], rownames=['Actual'], colnames=['Predicted'])

sns.heatmap(cm_rfc, annot=True, fmt= 'g')
plt.title('Confusion matrix for Random Forest Classifier')
ac_1 = accuracy_score(y_test, rfc_y_pred)
print('Accuracy is: ',ac_1)
plt.show()
```
